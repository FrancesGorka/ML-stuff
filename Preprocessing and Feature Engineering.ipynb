<span style="font-size: 20px;"> What is feature engineering? </span>

Feature engineering is a collection of approaches which prepare datasets prior to applying models to the data. This can include:
- removing or filling missing data
- scaling or normalising data
- feature selection
- creation of new features from old

1) Preliminary stage: Data preparation
- You might have to query databases, convert from csv or json files, put things into a pandas dataframe so you can analyse the data.

2) Exploratory data analysis (perform descriptive statistics on datasets).
- Look for correlated variables. Numpy has a np.corrcoef(x1,x2) to compare two vectors. <br>
<br>
- identify potential outliers using Z score, which is just a normalised data point. It's used to identify where on the normal distribution the value lies. If the z score is above 3 or below -3, then it may be an outlier.
- Check basic statistics (mean, median, variance, min, max) of numerical features. This tells you if the distribution is skewed, in which case you may want to perform a log transformation. Numpy has np.log(x). If features have low variance then they provide limited information about their influence on the target variable.
- Can look at pairwise relationships between features (bivarate analysis). Can do this via scatterplots.
- Can check normality of features via Shapiro-Wilk W test.The test is particularly useful for relatively small sample sizes (typically up to around 2000 observations)
- for categorical features, create frequency tables to understand distribution.
- analyze the distribution of the target variable for classification or regression tasks.Understand class imbalance in classification problems.

We can use this in the following way:
from scipy import stats
statistic, p_value = stats.shapiro(data)

3) Data cleaning
- handle missing values (dropping them or imputing them. Imputing can bias the data if the errors are systematic). 
- transforming dates 
- scaling, normalising

4) Feature engineering
- brainstorm new features 
- feature selection (find the most important features)

When you fit a linear regression model to a set of features, the model automatically assigns weights (coefficients) to each feature based on their contribution to explaining the variation in the target variable.

Decision trees will use identify feature importance. Once you've initialised a random forest and trained the model with model.fit, then depending on whether it's a classifier or a regression model, you can either use model.feature_importances_ to get an array of indices that correspond to important features OR mode.coef__ which identifies the weights. 

Apply PCA to the dataset before training the model: this will produce a new and uncorrelated feature set. It will reduce interpretability though.

- Recursive feature elimination
